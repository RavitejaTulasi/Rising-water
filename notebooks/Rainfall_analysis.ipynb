```python
# Fill missing values with mean

rainfall = rainfall.fillna(rainfall.mean(numeric_only=True))

print(rainfall.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

rainfall["STATE"] = le.fit_transform(rainfall["STATE"])

rainfall.head()
```

```python
rainfall.drop("COUNTRY", axis=1, inplace=True)

rainfall.head()
```

```python
rainfall.info()
```

```python
threshold = rainfall["ANNUAL"].quantile(0.75)

threshold
```

```python
rainfall["Flood"] = rainfall["ANNUAL"].apply(
    lambda x: 1 if x >= threshold else 0
)

rainfall.head()
```

```python
rainfall["Flood"].value_counts()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(x="Flood", data=rainfall)

plt.title("Flood Class Distribution")

plt.show()
```

```python
plt.figure(figsize=(14,10))

sns.heatmap(
    rainfall.corr(),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
X = rainfall.drop("Flood", axis=1)
y = rainfall["Flood"]

print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

```python
X = rainfall.drop(
    columns=[
        "Flood",
        "ANNUAL",
        "Jan-Feb",
        "Mar-May",
        "Jun-Sep",
        "Oct-Dec"
    ]
)

y = rainfall["Flood"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000)

lr.fit(X_train_scaled, y_train)
```

```python
y_pred_lr = lr.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred_lr)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_lr))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_lr)

print(cm)
```

```python
plt.figure(figsize=(6,5))

sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues"
)

plt.xlabel("Predicted")

plt.ylabel("Actual")

plt.title("Logistic Regression Confusion Matrix")

plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(
    random_state=42
)

dt.fit(X_train, y_train)
```

```python
y_pred_dt = dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

dt_accuracy = accuracy_score(y_test, y_pred_dt)

print("Decision Tree Accuracy:", dt_accuracy)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_dt))
```

```python
from sklearn.metrics import confusion_matrix

cm_dt = confusion_matrix(y_test, y_pred_dt)

print(cm_dt)
```

```python
plt.figure(figsize=(6,5))

sns.heatmap(
    cm_dt,
    annot=True,
    fmt="d",
    cmap="Greens"
)

plt.title("Decision Tree Confusion Matrix")

plt.xlabel("Predicted")

plt.ylabel("Actual")

plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf.fit(X_train, y_train)
```

```python
y_pred_rf = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

rf_accuracy = accuracy_score(y_test, y_pred_rf)

print("Random Forest Accuracy:", rf_accuracy)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_rf))
```

```python
from sklearn.metrics import confusion_matrix

cm_rf = confusion_matrix(y_test, y_pred_rf)

print(cm_rf)
```

```python
plt.figure(figsize=(6,5))

sns.heatmap(
    cm_rf,
    annot=True,
    fmt="d",
    cmap="Oranges"
)

plt.title("Random Forest Confusion Matrix")

plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.show()
```
